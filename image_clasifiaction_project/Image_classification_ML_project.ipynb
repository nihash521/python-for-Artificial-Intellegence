```python
#importing dependencies
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
#reading data set using pandas
data=pd.read_csv("C:/Users/Nihas/python-for-Artificial-Intellegence/image_clasifiaction_project/mnist.csv")
```

```python
#viewing column heads of the data
data.head()
```

```python
#extracting data set and viewing upto a close
a=data.iloc[3,1:].values
```

```python
#reshaping the data into reasonable size
a=a.reshape(28,28).astype('uint8')
plt.imshow(a)
```

```python
#preparing the data 
#seperating labels and data values
df_x=data.iloc[:,1:]
df_y=data.iloc[:,0]
```

```python
#creating train and test sizes/batches
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.2,random_state=4)
```

```python
#check data
x_train.head()
```

```python
#call random forest classifier
rf=RandomForestClassifier(n_estimators=100)
```

```python
#fit the model
rf.fit(x_train,y_train)
```

```python
#prediction on test data
pred=rf.predict(x_test)
```

```python
pred
```

```python
#check prediction accuracy
s=y_test.values
#caluculating number of correctly predicted values
count=0
for i in range(len(pred)):
    if pred[i]==s[i]:
        count=count+1
        
```
